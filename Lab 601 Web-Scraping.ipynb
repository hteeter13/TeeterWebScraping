```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://www.popvortex.com/music/charts/top-100-songs.php'
```

```python
scrapage = requests.get(url)
scrapage.status_code
```

```python
soup = BeautifulSoup(scrapage.content, "html.parser")
soup
```

```python
print(soup.prettify())
```

```python
print(soup.title)
```

```python
soup.select('#chart-position-1 > div.chart-content.col-xs-12.col-sm-8')
```

```python
print(soup.select('#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p'))
```

```python
soup.select("div.chart-content.col-xs-12.col-sm-8 > p > cite")[0].get_text()
```

```python
soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em")[0].get_text()
```

```python
song = []
artist = []

num_iter = len(soup.select("div.chart-content.col-xs-12.col-sm-8 > p > cite"))

for i in range(num_iter):
    song.append(soup.select("div.chart-content.col-xs-12.col-sm-8 > p > cite")[i].get_text())
    artist.append(soup.select("div.chart-content.col-xs-12.col-sm-8 > p > em")[i].get_text())
```

```python
print(song)
```

```python
print(artist)
```

```python
top100_songs = pd.DataFrame({"song":song,"artist":artist})
```

```python
top100_songs.head(10)
```

```python
top100_songs.to_csv('top100_songs.csv', index=False)
```
